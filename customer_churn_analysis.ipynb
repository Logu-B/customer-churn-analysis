```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/Telco-Customer-Churn.csv')

df.head()          # see first 5 rows
df.shape           # how many rows & columns
df.info()          # data types
df.isnull().sum()  # check missing values
df['Churn'].value_counts()  # how many churned vs not
```

```python
# Fix data types
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Fill or drop missing values
df.dropna(inplace=True)

# Remove unnecessary columns
df.drop('customerID', axis=1, inplace=True)
```

```python
import matplotlib.pyplot as plt

# Who churns more — monthly vs yearly contract?
plt.figure(figsize=(8, 5))
sns.countplot(x='Contract', hue='Churn', data=df)
plt.title('Churn Rate by Contract Type')
plt.xlabel('Contract Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Do high-paying customers churn more?
plt.figure(figsize=(8, 5))
sns.boxplot(x='Churn', y='MonthlyCharges', data=df)
plt.title('Monthly Charges vs. Churn')
plt.xlabel('Churn')
plt.ylabel('Monthly Charges')
plt.tight_layout()
plt.show()

# Tenure vs Churn
plt.figure(figsize=(8, 5))
sns.histplot(df[df['Churn']=='Yes']['tenure'], color='red', kde=True, bins=30)
plt.title('Tenure Distribution for Churned Customers')
plt.xlabel('Tenure (months)')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# At the end of your Python/Colab code, save the cleaned data
df.to_csv('cleaned_churn.csv', index=False)
```
